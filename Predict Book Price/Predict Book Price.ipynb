```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```
df = pd.read_csv("/content/drive/MyDrive/Practice Notebook/Predict Book Price/bestsellers with categories.csv")
df.head()
```

```
plt.scatter(df["Reviews"], df["Price"])
```

```
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, train_size = 0.8, test_size = 0.2, random_state = 0)
```

```
features = ['User Rating', 'Reviews', 'Year']
```

```
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(train[features].values, train["Price"])
```

```
prediction = model.predict(test[features])
labels = test['Price']

print(np.sqrt(sklearn.metrics.mean_squared_error(labels, prediction)))
```

```
plt.plot(test[features], test["Price"], ".",
         test[features], model.predict(test[features].values))
```
